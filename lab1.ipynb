### Задание 1

Определить (с обоснованием), зависимы или независимы следующие события:
<br> (a) Несовместные события;
<br> (b) События, образующие $σ$ -алгебру $Σ$ в пространстве $(Ω,Σ,P)$;
<br> (c) События, имеющие одинаковую вероятность?

#### Решение

События независимы, если
$$\mathsf{P}(AB)=\mathsf{P}(A)\mathsf{P}(B).$$


(a) Так как события несовместные, то $\mathsf{P}(AB)=0$.
Eсли $\mathsf{P}(A)=0$ или $\mathsf{P}(B)=0$, то события независимы $\mathsf{P}(AB)=\mathsf{P}(A)\mathsf{P}(B)=0$, иначе зависимы, так как $\mathsf{P}(A)\mathsf{P}(B)\neq0$.

(b) Рассмотрим эксперимент, заключающийся в подбрасывании симметричной монеты.

Простаранство элементарных исходов
$$\Omega=\{Р, О\}.$$

Сигма алгебра $Σ$
$$ Σ=\{\Omega, Р, О, \varnothing\}.$$

Пусть события $A$ - выпадение орла и $B$ - выпадение решки
$$\mathsf{P}(A)=\frac{1}{2}, \quad \mathsf{P}(B)=\frac{1}{2}, \quad \mathsf{P}(AB)=0.$$

События $A$ и $B$ являются зависимыми, так как $\mathsf{P}(A)\mathsf{P}(B)=\frac{1}{4}\neq \mathsf{P}(AB)$.

Пусть события $A$ - выпадение орла или решки и $B$ - выпадение решки
$$\mathsf{P}(A)=1, \quad \mathsf{P}(B)=\frac{1}{2}, \quad \mathsf{P}(AB)=\frac{1}{2}.$$

События $A$ и $B$ являются независимыми, так как $\mathsf{P}(A)\mathsf{P}(B)=\frac{1}{2} = \mathsf{P}(AB)$.

(c) Рассмотрим эксперимент, заключающийся в независимом подбрасывании двух симметричных монет.

Простаранство элементарных исходов
$$\Omega=\{РО, ОР, ОО, РР\}.$$

Пусть события $A$ - появление орла на первой монете, $B$ - появление орла на второй монете
$$A = \{ОР, ОО\}, \quad B = \{РО, ОО\}.$$

Вероятности событий $A$ и $B$
$$\mathsf{P}(A)=\mathsf{P}(B)=\frac{1}{2}⇒\mathsf{P}(A)\mathsf{P}(B)=\frac{1}{4}.$$

Вероятность события $AB$
$$\mathsf{P}(AB)=\frac{1}{4}.$$

Тогда $\mathsf{P}(AB)=\mathsf{P}(A)\mathsf{P}(B)=\frac{1}{4}⇒$ события $A$ и $B$ независимы.

Пусть события $A$ - появление хотя бы одного орла, $B$ - появление хотя бы одной решки
$$A = \{ОР, РО, ОО\}, B = \{ОР, РО, РР\}.$$

Вероятности событий $A$ и $B$
$$\mathsf{P}(A)=\mathsf{P}(B)=\frac{3}{4}⇒\mathsf{P}(A)\mathsf{P}(B)=\frac{9}{16}.$$

Вероятность события $AB$
$$\mathsf{P}(AB)=\frac{1}{2}.$$

Тогда $\mathsf{P}(AB)\neq\mathsf{P}(A)\mathsf{P}(B)⇒$ события $A$ и $B$ зависимы.

#### Решение

Если первый пароход подойдет к причалу первым в $x$ часов, тогда второму пароходу придется ожидать осбождение причала, если он подойдет в $y$ часов, где $x\leq y\leq x+1$. В случае если второй пароход подойдет к причалу первым в $y$ часов, тогда первому пароходу придется ожидать осбождение причала, если он подойдет в $x$ часов, где $y\leq x\leq y+2$.

Получаем, что $x-2\leq y \leq x+1$. Построим прямые $y=x+1$ и $y=x-2$ при $0\leq x \leq 24$, $0\leq y \leq 24$ и найдем площадь $S_3$ между ними:

$$S_3= S_{кв}-S_1-S_2=24^2-\frac{23^2}{2}-\frac{22^2}{2}=69.5$$

Тогда вероятность того, что одному из них придется ожидать освобождения причала
$$\mathsf{P}=\frac{S_3}{S_{кв}}=\frac{69.5}{24^2} \approx 0.12066$$

### Задание 5

Самолет, по которому ведется стрельба, состоит из трех различных по
уязвимости частей:
<br> (a) Кабина летчика и двигатель
<br> (b) Топливные баки
<br> (c) Планер

Для поражения самолета достаточно либо одного попадания в первую часть, либо двух попаданий во вторую, либо трех в третью. При попадании в самолет одного снаряда, снаряд с вероятностью $p_1$ попадает в первую часть, с вероятностью $p_2$ — во вторую, с вероятностью $p_3$ — в третью. Попавшие снаряды распределяются по частям независимо друг от друга.

Известно, что в самолет попало m снарядов. Найти условную вероятность $\mathsf{P}(A|m)$ события $A$ – «Самолет поражен» – при $m = 1, 2, 3, 4$.

#### Решение

Пусть события $B_1, B_2, B_3$ - попадание снаряда в первую, вторую и третью части самолета соответственно, $X_1, X_2, X_3$ - число попаданий снаряда в первую, вторую и третью части самолета соответственно.

1. При одном снаряде самолет будет поражен только при попадании снаряда в первую чать. Тогда

$$\mathsf{P}(A|m=1) = p_1.$$

2. В случае двух снарядов самолет не будет поражен, если один из снарядов попадет во вторую часть, а другой снаряд попадет в третью часть или оба снаряда попадут в третью часть. Тогда

$$\mathsf{P}(A|m=2) = 1-(\mathsf{P}(X_1=0, X_2=1, X_3=1)+\mathsf{P}(X_1=0, X_2=0, X_3=2))=1-\frac{2!}{1!1!}p_2p_3-\frac{2!}{2!}p_3^2=1-2p_2p_3-p_3^2.$$

3. В случае попадания 3 снарядов самолет не будет поражен, если один из снарядов попадет в третью часть, а другие два - во вторую. Тогда

$$\mathsf{P}(A|m=3) = 1-\mathsf{P}(X_1=0, X_2=1, X_3=2)=1-\frac{3!}{1!2!}p_2p_3^2 = 1-3p_2p_3^2.$$

4. При $m=4$ нет случаев, когда самолет не будет поражен. Если $X_1=0$, $X_2\leq 1$, $X_3\leq 2$, то $m\leq 3$, что противоречит условию. Тогда

$$\mathsf{P}(A|m=4) = 1.$$

### Задание 2

Совместное распределение случайных величин $ξ$ и $η$ задано следующей
таблицей
$$
\begin{array}{  c | c | c | c  }
\eta \backslash \xi & -1 & 0 & 1  \\ \hline
-1 & 1/8 & 1/12 & 7/24 \\ \hline
1 & 1/3 & 1/6 & 0
\end{array}
$$

(a) Найти маргинальные распределения $ξ$ и $η$
<br> (b) Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(ξ, η)$
<br> (c) Исследовать $ξ$ и $η$ на независимость и некоррелированность

#### Решение

(a) Составим таблицы маргинальных распределений  $ξ$ и $η$ :

$$
\begin{array}{  c | c | c | c  }
\eta & -1 & 0 & 1  \\ \hline
\mathsf{P} & 11/24 & 1/4 & 7/24
\end{array}
$$

$$
\begin{array}{  c | c | c }
\xi & -1 & 1  \\ \hline
\mathsf{P} & 1/2 & 1/2
\end{array}
$$

(b) Найдем соответствующие математические ожидания:

$$
\mathsf{E}\eta = -\frac{1}{6},
$$

$$
\mathsf{E}\xi = 0.
$$

Для составления ковариационной матрицы, вычислим ковариацию $\mathsf{cov}(\xi, \eta)$ и соответствующие дисперсии $\mathsf{D}\xi, \mathsf{D}\eta$ :

$$\mathsf{cov}(\xi, \eta) = \mathsf{E}(\xi\eta) - \mathsf{E}\xi\mathsf{E}\eta,$$

$$\mathsf{D}\xi=\mathsf{E}\xi^2-(\mathsf{E}\xi)^2, \mathsf{D}\eta=\mathsf{E}\eta^2-(\mathsf{E}\eta)^2.$$

Составим таблицу распределения для случайной величины $ξ^2, η^2, ξη$ :
$$
\begin{array}{  c | c }
\xi^2 & 1  \\ \hline
\mathsf{P} & 1
\end{array}
$$

$$
\begin{array}{  c | c| c }
\eta^2 & 0 & 1  \\ \hline
\mathsf{P} & 1/4 & 3/4
\end{array}
$$

$$
\begin{array}{  c | c| c| c }
\xi\eta & -1& 0 & 1  \\ \hline
\mathsf{P} & 15/24 & 1/4 & 1/8
\end{array}
$$
Тогда

$$\mathsf{cov}(\xi, \eta)=-\frac{15}{24}+\frac{1}{8}=-0.5,$$

$$\mathsf{D}\xi=1,$$

$$\mathsf{D}\eta=\frac{3}{4}-\frac{1}{36}=\frac{13}{18}.$$

Теперь составим ковариационную матрицу

$$
\begin{pmatrix}
\mathsf{D}\xi & \mathsf{cov}(\xi, \eta)\\
\mathsf{cov}(\xi, \eta) & \mathsf{D}\eta
\end{pmatrix}=\begin{pmatrix}
1 & -0.5\\
-0.5 & 13/18
\end{pmatrix}.
$$

Для составления матрицы корреляции найдем коэффициент корреляции $\rho$ для наших случайных величин:

$$
\rho(\xi,\eta) = \frac{\mathsf{cov}(\xi, \eta)}{\sqrt{\mathsf{D}\xi}\cdot\sqrt{\mathsf{D}\eta}} = \frac{-0.5}{1\sqrt{13/18}} \approx -0.588.
$$

Теперь составим корреляционную матрицу

$$
\begin{pmatrix}
1 & \rho(\xi,\eta)\\
\rho(\xi,\eta) & 1
\end{pmatrix}=\begin{pmatrix}
1 & -0.588\\
-0.588 & 1
\end{pmatrix}.
$$

(с) Так как коэффициент корреляции $\rho(\xi,\eta)\neq0 ⇒$ коррелированы.

Рассмотрим

$$\mathsf{P}(\xi=-1)\mathsf{P}(\eta=-1)=\frac{11}{24}\cdot\frac{1}{2}=\frac{11}{48}.$$

Тогда получаем, что $\mathsf{P}(\xi=-1,\eta=-1)=\frac{1}{8} \neq \frac{11}{48} ⇒$ зависимы.

#### Решение

Распределение равномерно абсолютно непрерывной случайной величины $\xi$ задано функцией плотности вероятности $f_ξ(x)$:

$$f_{\xi}(x) =  \begin{cases} \frac{1}{2\pi}, & x \in [-\pi,\pi] \\ 0, & x \not\in [-\pi,\pi] \end{cases}.$$

(a) Вычислим математические ожидания случайных величин $\eta_1, \eta_2$ и их квадратов:

$$\mathsf{E}\eta_1 = \int\limits_{-\pi}^{\pi}cos(x)\cdot\frac{1}{2\pi}dx=\frac{1}{2\pi}(sin(\pi)-sin(-\pi))=0$$

$$\mathsf{E}\eta_1^2 = \int\limits_{-\pi}^{\pi}cos^2(x)\cdot\frac{1}{2\pi}dx=\frac{1}{2\pi}\int\limits_{-\pi}^{\pi}\Bigl(\frac{1}{2}+\frac{cos2x}{2}\Bigl)dx=\frac{1}{2\pi}\Bigl(\frac{x}{2}+\frac{sin2x}{4}\Bigl)\Bigg|_{-\pi}^{\pi}=\frac{1}{4}+\frac{1}{4}=\frac{1}{2}$$

$$\mathsf{E}\eta_2 = \int\limits_{-\pi}^{\pi}sin(x)\cdot\frac{1}{2\pi}dx=\frac{1}{2\pi}(-cos(\pi)+cos(-\pi))=0$$

$$\mathsf{E}\eta_2^2 = E(1-\eta_1^2)=1-E\eta_1^2=\frac{1}{2}$$

Вычислим дисперсии случайных величин $\eta_1$ и $\eta_2$:

$$D\eta_1 = \mathsf{E}\eta_1^2-(\mathsf{E}\eta_1)^2 =\frac{1}{2} $$

$$D\eta_2 = \mathsf{E}\eta_2^2-(\mathsf{E}\eta_2)^2 =\frac{1}{2} $$

Определим ковариацию $\mathsf{cov}(\eta_1, \eta_2)$ :
$$\mathsf{E}(\eta_1\eta_2)=\int\limits_{-\pi}^{\pi}\frac{1}{2\pi}cos(x)sin(x) dx = 0$$

Тогда
$$\mathsf{cov}(\eta_1, \eta_2) = \mathsf{E}(\eta_1\eta_2)-\mathsf{E}(\eta_1)\mathsf{E}(\eta_2)=0$$

Составим ковариационную матрицу:

$$
\begin{pmatrix}
\mathsf{D}\eta_1 & \mathsf{cov}(\eta_1, \eta_2)\\
\mathsf{cov}(\eta_1, \eta_2) & \mathsf{D}\eta_2
\end{pmatrix}=\begin{pmatrix}
0.5 & 0\\
0 & 0.5
\end{pmatrix}.
$$

Составим корреляционную матрицу:

$$
\begin{pmatrix}
1 & \rho(\eta_1,\eta_2)\\
\rho(\eta_1,\eta_2) & 1
\end{pmatrix}=\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}.
$$

(b)
Так как коэффициент корреляции  $\mathsf{\rho}(\eta_1,\eta_2)=0 ⇒$  некоррелированы.

Поскольку $\eta_1^2+\eta_2^2=cos^2\xi+sin^2\xi=1$, то $\eta_1$ и $\eta_2$ зависимы.

### Задание 5

Найти плотность распределения суммы двух независимых случайных величин $ξ$ и $η$, если $ξ\sim Exp_2$ и $η \sim U_{0,1}$.

#### Решение

Так как
$$
f_{\xi}(x) = \begin{cases} 0, & x < 0 \\ 2e^{-2x}, & x\geq 0 \end{cases}.
$$
и
$$
f_{\eta}(x) = \begin{cases} 0, & x \notin [0,1] \\ 1, & x \in [0,1] \end{cases},
$$
то
$$
f_{\xi+\eta}(x) = \int\limits_{0}^\infty 2e^{-2t}f_{\eta}(x-t)\,dt.
$$

Перед нами полоса
$$t\geq 0, \quad 0 \leq x-t \leq 1.$$

1. Если $x \in [0,1]$, то $t \in [0, x]$
$$
f_{\xi+\eta}(x) = \int\limits_{0}^x 2e^{-2t}dt=-e^{-2x}+1.
$$

2. Если $x > 1$, то $t \in [x-1, x]$
$$
f_{\xi+\eta}(x) = \int\limits_{x-1}^x 2e^{-2t}dt=-e^{-2x}+e^{-2(x-1)}.
$$

Тогда плотность распределения суммы двух независимых случайных величин $ξ$ и $η$ :

$$
f_{\xi+\eta}(x) = \begin{cases} 1-e^{-2x}, & 0 \leq x \leq 1 \\ e^{2-2x}-e^{-2x}, & x > 1 \end{cases}.
$$
